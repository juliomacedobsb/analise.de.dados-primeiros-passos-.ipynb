```python
#importando a base de dados para vizualização



import pandas as pd

tabela = pd.read_excel("Vendas.xlsx")
tabela.head()
```

```python
# analisando o faturamento da empresa


faturamento_total = tabela['Valor Final'].sum()
faturamento_total
```

```python
#  analisando faturamento por loja
faturamento_loja = tabela[['ID Loja', 'Valor Final' ]].groupby('ID Loja').sum()
faturamento_loja.head()

```

```python
# observando produtos de cada loja e análisando possíveis motivos para maior faturamento

faturamento_por_produto = tabela[['ID Loja', 'Produto', 'Valor Final']].groupby(['ID Loja', "Produto"]).sum()
faturamento_por_produto

#observando q a bermuda lisa e o produto mais vendido na loja de uberlandia e que seria possível colocar esse produto nas outras lojas.
```
